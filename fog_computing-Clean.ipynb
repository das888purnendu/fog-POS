```python
import random
import numpy as np 
```

FOG particles and connections

```python
total_node=10
cloud=1
fog=4
edge=2
iot=total_node-(cloud-fog-edge)
fog_properties={i:
                {"power_consumption":round(random.randint(1,10)*random.random(),2),
                 "storage_capacity":round(random.randint(1,10)*random.random(),2),
                 "load_capacity":round(random.randint(1,10)*random.random(),2),
                "position_vector":np.array([round(random.randint(-50,50) * random.random(),3),
                                   round(random.randint(-50,50) * random.random(),3)])}
                for i in range(1,fog+1)}
```

```python
connections=[[0 for j in range (total_node)] for i in range (total_node)]

ca=0
cb=0
numa=0
numb=0


devices=["Cloud_node","Fog_node","Edge_node","IOT_node"]

for i in range(total_node):
    if(i<cloud):
        ca=0
        numa=i+1
    elif(i<cloud+fog):
        ca=1
        numa=(i-cloud)+1
    elif(i<cloud+fog+edge):
        ca=2
        numa=(i-(cloud+fog))+1
    else:
        ca=3
        numa=(i-(cloud+fog+edge))+1
    cda=devices[ca]
    
    
    for j in range(total_node):

        if(j<cloud):
            cb=0
            numb=j+1
        elif(j<cloud+fog):
            cb=1
            numb=(j-cloud)+1
        elif(j<cloud+fog+edge):
            cb=2
            numb=(j-(cloud+fog))+1
        else:
            cb=3
            numb=(j-(cloud+fog+edge))+1
        cdb=devices[cb]
        if(j==i):
            continue
        elif(j>i):
            connections[i][j]=int(input("{}-{} and {}-{} : ".format(cda,numa,cdb,numb)))
        else:
            connections[i][j]=connections[j][i]
    print("\n")
```

```python
print("Connection Matrix -")
for i in connections:
    print(*i)
```

```python
print("Fog nodes properties -\n")
for i in fog_properties.items():
    print("Fog Node "+str(i[0]))
    print("\
        power_consumption : {}\n\
        storage_capacity :{}\n\
        load_capacity : {}\n\
        position_vector : {},{}".format(i[1]["power_consumption"]
                                        ,i[1]["storage_capacity"],
                                        i[1]["load_capacity"],
                                        i[1]["position_vector"][0],
                                        i[1]["position_vector"][1]))
    print("")
```

POS algo -

```python
def fitness_function(position):
    return position[0]**2 + position[1]**2 + 1


W = 0.5
c1 = 0.5
c2 = 0.9
target = 1

n_iterations = 10
target_error = 5
n_particles = fog #number of Fog nodes


# Fetching the fog positions from the fog properties -
particle_position_vector =np.array([ i[1]["position_vector"] for i in fog_properties.items()])
pbest_position = particle_position_vector

pbest_fitness_value = np.array([float('inf') for _ in range(n_particles)])
gbest_fitness_value = float('inf')
gbest_position = np.array([float('inf'), float('inf')])

velocity_vector = ([np.array([0, 0]) for _ in range(n_particles)])

for iteration in range(n_iterations):
    for i in range(n_particles):
        fitness_cadidate = fitness_function(particle_position_vector[i])
        
        if(pbest_fitness_value[i] > fitness_cadidate):
            pbest_fitness_value[i] = fitness_cadidate
            pbest_position[i] = particle_position_vector[i]

        if(gbest_fitness_value > fitness_cadidate):
            gbest_fitness_value = fitness_cadidate
            gbest_position = particle_position_vector[i]

    if(abs(gbest_fitness_value - target) < target_error):
        break
    
    for i in range(n_particles):
        new_velocity = (W*velocity_vector[i]) + (c1*random.random()) * (pbest_position[i] - particle_position_vector[i]) + (c2*random.random()) * (gbest_position-particle_position_vector[i])
        new_position = new_velocity + particle_position_vector[i]
        particle_position_vector[i] = new_position

    
print("The best position is ", gbest_position, "in iteration number ", iteration)
```
